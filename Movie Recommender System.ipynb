```python
#reading data
df1=pd.read_csv('movies.csv')
df2=pd.read_csv('ratings.csv')
```

```python
df1.head()
```

```python
df2.head()
```

```python
#changing df1 
new_df1 = df1.drop('genres', axis=1).join(df1['genres'].str.get_dummies(sep='|'))
new_df1
```

```python
#merge new_df1 and df2 on movieId
merged_df = pd.merge(df2, new_df1, on='movieId')

#dropping the no genres listed column
final_df = merged_df.drop('(no genres listed)', axis = 1)
final_df.head()
```

```python
#dropping the no genres listed column and setting movieId as index
new_df1 = new_df1.drop('(no genres listed)', axis = 1)
new_df1 = new_df1.set_index('movieId')
new_df1.head()
```

```python
#Item-Feature Matrix

item_features = new_df1.drop('title', axis = 1)

item_features
```

```python
#Normalized Item-Feature Matrix

item_features_norm = item_features.divide(item_features.sum(axis = 0))

item_features_norm.head()
```

```python
#dropping movieId, timestamp and title from final_df
droped_final_df = final_df.drop(columns = ['movieId', 'timestamp', 'title'])

droped_final_df
```

```python
#multiplying the genre columns with the corresponding ratings
for i in range(2,21):
    droped_final_df.iloc[:,i] = droped_final_df.iloc[:,i]*droped_final_df['rating']
    
#dropping the rating column
droped_final_df = droped_final_df.drop(columns = "rating")

droped_final_df
```

```python
#grouping by userId and adding up
per_user = droped_final_df.groupby('userId').sum()

per_user
```

```python
#getting the sum for each user
sum_per_user = per_user.sum(axis=1)

sum_per_user
```

```python
#user-feature matrix

user_features = per_user.divide(sum_per_user,axis=0)

user_features
```

```python
#now I will pick up a random user and recommend some movies for him/her using cosine similarity. Let say, user = 610
```

```python
user = 610
similarity = pd.DataFrame(item_features_norm.dot(user_features.loc[user].T))
similarity.columns = ['sim']

#recommending 10 movies to user 610
similarity.sort_values('sim', ascending = False).head(10).merge(df1.set_index('movieId'), left_index=True, right_index=True)
```

```python
#now finding out how many movies the user has rated among our recommendation and checking the sensibility of our recommendation
df2_610 = df2[df2['userId'] == 610]
df2_610.loc[df2_610['movieId'].isin([81132, 79132, 60684, 26701, 43932, 56174, 80363, 83349, 85261, 88125])]
```

```python
#now see what's the average rating given by user 610
df2_610['rating'].mean()
```

```python
#the average rating is 3.6. So, our recommendation seems sensible. Now, let's take another user. 
```

```python
#here we can't see the RMSE value of MAE value to check the performance of the system because we used content based approach.
#Thus, I choosed two users randomly to see the performance of the system. Now, we will see model based approach and we will 
#we will check the performance using RMSE and MAE value. Let's jump into it.
```

```python
from surprise import Reader, Dataset
from surprise.model_selection import train_test_split
from surprise import accuracy
from surprise import SVD
from surprise import KNNBasic
from surprise.model_selection import GridSearchCV
```

```python
import numpy as np
```

```python
#again reading the data and transforming it in a usable manner
ratings = pd.read_csv('ratings.csv')
ratings_dict = {'itemID': list(ratings.movieId),
                'userID': list(ratings.userId),
                'rating': list(ratings.rating)}
df = pd.DataFrame(ratings_dict)
```

```python
# A reader is still needed but only the rating_scale param is requiered
reader = Reader(rating_scale=(0, 5.0))
```

```python
# The columns must correspond to user id, item id and ratings (in that order)
data = Dataset.load_from_df(df[['userID', 'itemID', 'rating']], reader)
```

```python
#train test split
trainset, testset = train_test_split(data, test_size=.25)
```

```python
#first let's work with single value decomposition
algo = SVD()
```

```python
# Train the algorithm on the trainset, and predict ratings for the testset
algo.fit(trainset)
predictions = algo.test(testset)
predictions
```

```python
#from the above predictions we can manually see from the r_ui value and est value the performance of the system but we will
#look into the RMSE and MAE value to check. 
```

```python
accuracy.rmse(predictions)
```

```python
accuracy.mae(predictions)
```

```python
#From the RMSE and MAE, it can be told that it is working well with our test set. Now, let's jump into KNNBasic and compare
#the performace of KNNBasic and SVD.
```

```python
algo = KNNBasic()
algo.fit(trainset)
predictions = algo.test(testset)
predictions
```

```python
accuracy.rmse(predictions)
```

```python
accuracy.mae(predictions)
```

```python
#Clearly, SVD is working well with our test data. Hence, I am choosing SVD and I will tune the hyperparameters now to improve 
#the RMSE AND MAE
```

```python
#specifying the parameter values
param_grid = {'n_epochs': [5, 10], 'lr_all': [0.002, 0.005],
              'reg_all': [0.4, 0.6]}

gs = GridSearchCV(SVD, param_grid, measures=['rmse', 'mae'], cv=5)

gs.fit(data)

# best RMSE score
print(gs.best_score['rmse'])

# combination of parameters that gave the best RMSE score
print(gs.best_params['rmse'])
```

```python
# best MAE score
print(gs.best_score['mae'])

# combination of parameters that gave the best MAE score
print(gs.best_params['mae'])
```

```python
#we can't see a improvement.
```

```python
#trying a different number of cv
gs = GridSearchCV(SVD, param_grid, measures=['rmse', 'mae'], cv=3)

gs.fit(data)

# best RMSE score
print(gs.best_score['rmse'])

# combination of parameters that gave the best RMSE score
print(gs.best_params['rmse'])
```

```python
# best MAE score
print(gs.best_score['mae'])

# combination of parameters that gave the best MAE score
print(gs.best_params['mae'])
```

```python
#still, no improvement. To conclude, it can be told that among SVD and KNNBasic, SVD performed well with our test set
#and the content based approach also looks sensible.
```
